```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
# from statsmodels.sandbox.distributions.multivariate import multivariate_t_rvs # for t-distribution

random.seed(12)
```

```python
def width2var(w):
    var = (w/2)**2
    return var

# Choose roughly 60% of the intended width as width2var param
```
